```python
#fetching data from csv file
df = pd.read_csv('Sheet 1.csv')
df.info()
```

```python
states = df ['State'].tolist()
crime = df ['Crime'].tolist()
well_being = df ['Well-being'].tolist()

plt.plot(states, crime, marker='o', linewidth=3, label='Crime')
plt.plot(states, well_being, marker='o', linewidth=3, label='Well being')
plt.xlim(0,7)
plt.xlabel("States")
plt.ylabel("Rate")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Assuming you have your data in 'states', 'crime', and 'well_being' lists
states = df['State'].tolist()
crime = df['Crime'].tolist()
well_being = df['Well-being'].tolist()

# Calculate the number of plots and states per plot
num_plots = 10
states_per_plot = len(states) // num_plots  # Assuming it's evenly divisible

# Create subplots
fig, axs = plt.subplots(num_plots, 1, figsize=(10, 20))  # Adjust the figsize as needed

# Loop through and create each subplot
for i in range(num_plots):
    start_idx = i * states_per_plot
    end_idx = start_idx + states_per_plot

    axs[i].plot(states[start_idx:end_idx], crime[start_idx:end_idx], marker='o', linewidth=3, label='Crime')
    axs[i].plot(states[start_idx:end_idx], well_being[start_idx:end_idx], marker='o', linewidth=3, label='Well-being')
    axs[i].set_title(f"States {start_idx+1} to {end_idx}")

    axs[i].set_xlabel("States")
    axs[i].set_ylabel("Rate")
    axs[i].grid(True)

# Adjust the layout to prevent overlapping labels
plt.tight_layout()

# Display the plots
plt.show()
```
